```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on= 'Country Code', how= 'left')
```

```python
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 countries

plt.pie(country_val[:3], labels= country_names[:3], autopct='%1.2f%%')
```

## Observation

Zomato maximum records or transactions are from India, After that USA and then United Kingdom

```python
final_df.columns
```

```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings
```

## Observations

- When Ratings are from 4.5 to 4.9 -----> Excellent
- When Ratins are from 4.0 to 4.4 -------> Very Good
- When Ratings are from 3.5 to 3.9 ------> Good
- When Ratings are from 2.5 to 3.4 ------> Average
- When Ratings are from 1.8 to 2.4 -------> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',hue= 'Rating color', data=Ratings, palette=['blue','red','orange','yellow','green', 'green'])
```

## Observation

- Not rated count is very high
- Maximum number of rating are between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x='Rating color', data=Ratings, palette=['blue','red','orange','yellow','green','green'])
```

```python
## Find the name of countries that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

## Observations

Maximum number of 0 ratings are from India

```python
## Find out which currency is used by which country?
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
## Which countries has online deliveries option?

final_df[final_df['Has Online delivery']=='Yes']. Country.value_counts()
```

```python
final_df[['Has Online delivery', 'Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```

## Observation

```python
## Create a pie chart for top 5 cities

city_values= final_df.City.value_counts().values
city_lables= final_df.City.value_counts().index

plt.pie(city_values[:5], labels=city_lables[:5], autopct='%1.2f%%')
```

```python
## Find top 10 cuisines

top_10_cuisine= final_df['Cuisines'].str.split(',').explode().str.strip().value_counts().head(10).reset_index()
top_10_cuisine.columns= ['Cuisines','Count']

plt.figure(figsize=(12,6))
sns.barplot(x='Cuisines',y='Count', data= top_10_cuisine, palette='viridis', edgecolor='black')
plt.title('Top 10 Most Frequent Cuisines', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Cuisine Type', fontsize=12)
plt.ylabel('Count of Restaurant listing the cuisine', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
